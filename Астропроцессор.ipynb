```
from flatlib import const
from flatlib.chart import Chart
from flatlib.datetime import Datetime
from flatlib.geopos import GeoPos
from flatlib import aspects
```

```
date = Datetime('1976/08/01', '6:30', '+05:00')
pos = GeoPos(42.8767897,74.4517772)
chart = Chart(date, pos, IDs=const.LIST_OBJECTS)
```

```
sun = chart.get(const.SUN)
print(sun)
```

```
house = chart.houses.getObjectHouse(sun)
print(house)
```

```
for planet in const.LIST_OBJECTS:
  aspect = aspects.getAspect(sun, chart.get(planet), const.MAJOR_ASPECTS)
  print(aspect) if aspect.exists() else False
```
